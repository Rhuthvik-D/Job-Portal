```python
# %pip install playwright

# Install Playwright browsers (run this cell once)
!playwright install

import asyncio
from playwright.async_api import async_playwright

async def scrape_jobs():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        await page.goto("https://wellfound.com/role/sales-development", timeout=60000)
        await page.wait_for_timeout(5000)  # let JS load

        # Example: Get job titles
        jobs = await page.locator("a.mr-2.text-sm.font-semibold.text-brand-burgandy").all()
        for job in jobs:
            print(await job.inner_text(), "|", await job.get_attribute("href"))
        await browser.close()

```
