```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# Setup headless Chrome
options = Options()
options.add_argument('--headless=new')
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64)")
driver = webdriver.Chrome(options=options)
```

```python
# Go to Naukri SDR jobs page
driver.get("https://www.naukri.com/sdr-jobs?k=sdr")
```

```python
WebDriverWait(driver, 15).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a.title')))

a_tags = driver.find_elements(By.CSS_SELECTOR, 'a.title')

job_weblinks = []

for a in a_tags:
    if a.get_attribute("href") is not None:
        job_weblinks.append(a.get_attribute("href"))
```

```python
WebDriverWait(driver, 15).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'span.comp-dtls-wrap')))

a_tags = driver.find_elements(By.CSS_SELECTOR, 'span.comp-dtls-wrap')

company_names = []

for a in a_tags:
    if len(a.text.strip())!=0:
        company_names.append(a.text.strip().split('\n')[0])
    else:
        pass
```

```python
job_title = []
description = []
experience = []
salary = []
location = []

for each_url in job_weblinks:
    options = Options()
    options.add_argument('--headless=new')
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64)")
    driver = webdriver.Chrome(options=options)

    print(f"Visiting: {each_url}")
    driver.get(each_url)

    # 1. Extract h1 text
    try:
        header = driver.find_element(By.XPATH, "//h1[contains(@class, 'styles_jd-header-title__rZwM1')]")
        job_title_text = header.text.strip()
        print(job_title_text)
        job_title.append(job_title_text)
    except:
        job_title.append(0)
        print("❌ H1 with that class not found")

    try:
        header = driver.find_element(By.CLASS_NAME, "styles_job-desc-container__txpYf")
        print(header.text.strip())
        description.append(header.text.strip())
    except:
        description.append(0)
        # print("❌ Section Title with that class not found")

    try:
        header = driver.find_element(By.CLASS_NAME, "styles_jhc__exp__k_giM")
        print(header.text.strip())
        experience.append(header.text.strip())  # Assuming the first line is the job title
    except:
        experience.append(0)
        # print("❌ H1 with that class not found")

    try:
        header = driver.find_element(By.CLASS_NAME, "styles_jhc__salary__jdfEC")
        print(header.text.strip())
        salary.append(header.text.strip())  # Assuming the first line is the job title
    except:
        salary.append(0)
        # print("❌ H1 with that class not found")

    try:
        header = driver.find_element(By.CLASS_NAME, "styles_jhc__loc___Du2H")
        print(header.text.strip())
        location.append(header.text.strip())  # Assuming the first line is the job title
    except:
        location.append(0)
        # print("❌ H1 with that class not found")
```
